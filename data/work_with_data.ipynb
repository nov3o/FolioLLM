```python
import pandas as pd
import numpy as np
import timeit

# Save the DataFrame in different formats
df.to_csv('sample.csv', index=False)
df.to_pickle('sample.pkl')
df.to_excel('sample.xlsx', index=False)
df.to_parquet('sample.parquet')
df.to_feather('sample.feather')

# Redefine the read functions
def read_csv():
    pd.read_csv('sample.csv')

def read_pickle():
    pd.read_pickle('sample.pkl')

def read_excel():
    pd.read_excel('sample.xlsx')

def read_parquet():
    pd.read_parquet('sample.parquet')

def read_feather():
    pd.read_feather('sample.feather')

# Measure the read times for CSV, Pickle, Excel, Parquet, and Feather formats
csv_time = timeit.timeit(read_csv, number=5)
pickle_time = timeit.timeit(read_pickle, number=5)
excel_time = timeit.timeit(read_excel, number=5)
parquet_time = timeit.timeit(read_parquet, number=5)
feather_time = timeit.timeit(read_feather, number=5)

csv_time, pickle_time, excel_time, parquet_time, feather_time
```
